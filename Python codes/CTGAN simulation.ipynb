```python
# Define your Python function
def my_function(input_data):
    #Realdata = pd.read_csv('Name of your Dataset')
    Realdata =pd.read_csv(input_data)


# Create metadata (adjust for your dataset's structure)
    metadata = SingleTableMetadata()
    metadata.detect_from_dataframe(Realdata)

# Identify numerical and categorical columns
    num_cols = Realdata.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_cols = Realdata.select_dtypes(include=['object', 'category']).columns.tolist()


    model = CTGANSynthesizer(
         metadata, # required
         enforce_rounding=False,
         #epochs=3000,
         epochs=1000,
          #batch_size=128,
         verbose=False)
# Fit the model to your dataset
    model.fit(Realdata)
# Sample synthetic data
    random.seed(202403)
    synth_data = model.sample(num_rows=1000)
    return synth_data
```

```python
import pandas as pd
import zipfile
import os
import time

# Record the start time
start_time = time.time()
# Define the directory to save CSV files
csv_dir = '/content/'
# List of inputs for the function
#inputs = ['veteran1.csv']
#inputs = ['veteran2.csv']
inputs = ['veteran3.csv']

num_repeats = 30

# Run the function for each input and save results as CSV files
for i, input_value in enumerate(inputs):
  for j in range(num_repeats):
    # Call the function
    result = my_function(input_value)

    # Convert the result to a DataFrame (assuming it's a single value)
    df_result = result

    # Save the DataFrame as a CSV file
    file_name = f'veteran3batch3_sim{i}_{j}.csv'
    df_result.to_csv(os.path.join(csv_dir, file_name), index=False)

# Create a zip file containing all CSV files
zip_file_name = '/content/csv_files.zip'
with zipfile.ZipFile(zip_file_name, 'w') as zipf:
    # Add each CSV file to the zip file
    for i in range(len(inputs)):
        for j in range(num_repeats):
            file_name = f'veteran3batch3_sim{i}_{j}.csv'
            zipf.write(os.path.join(csv_dir, file_name), file_name)

# Download the zip file
from google.colab import files
files.download(zip_file_name)

# Record the end time
end_time = time.time()

# Calculate the runtime
runtime = end_time - start_time
print("Runtime:", runtime, "seconds")
```
